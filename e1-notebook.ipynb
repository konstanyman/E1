<h2>3.</h2>

```python
# dataframe (13-data.csv)

df1 = P.read_csv('./e1-data/13-data.csv',
            # field separator character
            sep=",",
            # missing value characters
            na_values=["?"],
            # no row names
            index_col=False,
            # column names
            header=0,
            # quote character
            quotechar="'")
df1 = df1.replace("\"", "", regex=True)
df1.columns = df1.columns.str.replace("\"", "")
df1
```

<h3>3.1.</h3>

<h3>Find invalid values</h3>

```python
# column 'sex'

count = 0
for x in df1['sex']:
    count += 1
    if x != 'male' and x != 'female':
        print(x, count)

df1 = df1.replace('nale', 'male')
print(df1.iloc[61,0])
```

```python
# column 'location'

for x in df1['location']:
    if x != 'Turku' and x != 'Helsinki' and x != 'Tampere':
        print(x)
    
df1 = df1.replace('33100', 'Tampere')
df1 = df1.replace('20100', 'Turku')
```

```python
# column 'purchases'

for x in df1['purchases']:
    if x < 0 or x > 15 or isinstance(x, float) == False:
        print(x)
```

```python
# column 'retention_time'

for x in df1['retention_time']:
    if x < 0 or isinstance(x, float) == False:
        print(x)
df1['retention_time'] = abs(df1['retention_time'])
```

<h3>3.2.</h3>

```python
# replace all missing values of the "purchases" variable with zero

df1["purchases"] = df1["purchases"].fillna(0)
df1.info()
    
```

<h3>3.3.</h3>

```python
# grouping

grouped = df1.groupby(['sex', 'location'])
grouped.first()
```

```python
# replace NaN values with corresponding medians from each group

def replace_with_group_median(group):
    median = group['retention_time'].median()
    group['retention_time'].fillna(median, inplace=True)
    return group

df1 = grouped.apply(replace_with_group_median).reset_index(drop=True)
df1
```

<h2>4.</h2>

```python
# load files

# espoo

df_e = P.read_csv('./e1-data/14-espoo.csv',
            # field separator character
            sep=",",
            # missing value characters
            na_values=["?"],
            # no row names
            index_col=False,
            # column names
            header=0,
            # quote character
            quotechar="'")

# helsinki

df_h = P.read_csv('./e1-data/14-helsinki.csv',
            # field separator character
            sep=",",
            # missing value characters
            na_values=["?"],
            # no row names
            index_col=False,
            # column names
            header=0,
            # quote character
            quotechar="'")
```

```python
# merge dataframes

df_eh = P.merge(df_e, df_h, on='date', how='outer')
df_eh
```

<h3>4.1.<h3>

```python
# check if whole row is NaN

def some_rows_all_NaN(df):
    nan_rows = df.isna().all(axis=1)
    if nan_rows.any() == False:
        print('There are no rows in dataframe where all values are NaN')
        return(False)
    else:
        print('There are some rows in dataframe where all values are NaN')
        return(True)

some_rows_all_NaN(df_eh)
```

Observations are made in 3134 days in total.

<h3>4.2.<h3>

```python
# counter for non-NaN observations each column

def ob_per_column(df):
    return(df.count())

ob_per_column(df_eh)
```

<h3>4.3.<h3>

```python
# counter for rows where all values are non-NaN

def full_row_counter(df):
    return(df.notna().all(axis=1).sum())

print('There were', full_row_counter(df_eh), 'days where all streets were observed.')
```

<h3>4.4.<h3>

```python
# busiest street

def total_count(df):
    totals = df.sum(numeric_only=True)
    sorted_totals = totals.sort_values(ascending=False)
    return(sorted_totals)

print(f'Order from busiest to least busy streets:\n\n{total_count(df_eh)}')
```

<h3>4.5.<h3>

```python
# actual busiest street

def real_total_count(df):
    real_df = df.dropna()
    totals = real_df.sum(numeric_only=True)
    sorted_totals = totals.sort_values(ascending=False)
    return(sorted_totals)

print(f'Actual order from busiest to least busy streets:\n\n{real_total_count(df_eh)}')
```

Filtering out rows where some values are NaN of course affects the conclusion. Some streets were observed more which the first calculation doesn't take in consideration. The busyness of the street is the average amount of cyclists per day.
