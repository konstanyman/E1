```python
# imports

import csv as C
import numpy as N
import pandas as P
```

```python
# names

names = P.read_csv('./e1-data/thyroid+disease/allbp.names',
            # field separator character
            sep="\t",
            # missing value characters
            na_values=["?"],
            # no row names
            index_col=False,
            # column names
            header=None,
            # quote character
            quotechar="'",
            # skip unneccessary rows
            skiprows=9)
names_last = P.DataFrame([['protein:']])
names_full = P.concat([names, names_last]) 
names_full
```

```python
# dataframe

df = P.read_csv('./e1-data/thyroid+disease/allbp.data',
            # field separator character
            sep=",",
            # missing value characters
            na_values=["?"],
            # no row names
            index_col=False,
            # column names
            header=None,
            # quote character
            quotechar="'")
df.columns = [names_full.iloc[:,0]]
df
```

```python
# NaN counter

i = 0
while i < 30:
    count = 0
    for x in df.iloc[:,i]:
        if P.isna(x):
            count += 1
    print("Variable", list(df.columns)[i], "contains", count, "missing values")
    i += 1
```


1. There are 2800 observations with 30 variables
2. There are missing values in 8 of the variables. See "NaN counter"

<h3>2.1.</h3>

```python
# yes/no counter

i = 0
while i < 30:
    if df.iloc[1,i] == "f" or df.iloc[1,i] == "t":
        count = 0
        for x in df.iloc[:,i]:
            if x == 't':
                count += 1
        division = round(count/2800, 2)
        print('Percentage of yes-values in variable', list(df.columns)[i], 'is', division, '%')
    i += 1
```

<h3>2.2.</h3>

```python
# for each of the TSH, T3, TT4, T4U, FTI and TBG variables, calculate the sum of the squared values divided by the number of non-NA values

for i in range(17,28,2):
    count = 0
    nancount = 0
    for x in df.iloc[:,i]:
        if P.isna(x):
            nancount += 1
            continue
        count = count + x**2
    print(list(df.columns)[i], round(count, 2), "/", nancount, "=", round(count/nancount, 2))

        
```

<h3>2.3.</h3>

```python
# calculate the mean ratio between T3 and TT4 (if mean ratio the ratio of means)

means = df[['T3:', 'TT4:']].mean()

ratio = round(means[0]/means[1], 4)

print("Ratio between the means of T3 and TT4 is", ratio)
```
